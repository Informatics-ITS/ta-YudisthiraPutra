```python
import textdistance

# Compute similarity scores
def compute_similarity(s1, s2):
    return {
        'jaccard': textdistance.jaccard(s1, s2),
        'sorensen_dice': textdistance.sorensen(s1, s2),
        'jaro': textdistance.jaro(s1, s2),
        'jaro_winkler': textdistance.jaro_winkler(s1, s2),
        'ratcliff_obershelp': textdistance.ratcliff_obershelp(s1, s2)
    }

# Cluster characters without pointers
def cluster_without_pointers(characters_list, aliases_clusters, threshold):
    cluster_id = len(aliases_clusters) + 1
    for character in characters_list:
        found = False
        for key, cluster in aliases_clusters.items():
            for name in cluster:
                if character.endswith("nya") and name.endswith("nya"):
                    # print(character)
                    # print(name)
                    if compute_similarity(character[:-3].lower(), name[:-3].lower())['jaro_winkler'] >= threshold:
                        aliases_clusters[key].add(character)
                        found = True
                        break
                    else:
                        break
                if character.lower() in name.lower() or compute_similarity(character.lower(), name.lower())['jaro_winkler'] >= threshold:
                    aliases_clusters[key].add(character)
                    found = True
                    break
            if found:
                break
        if not found:
            aliases_clusters[f"person-{cluster_id}"] = {character}
            cluster_id += 1
    return aliases_clusters

# Cluster characters with pointers
def cluster_with_pointers(characters_with_pointer, pointers, threshold):
    all_pointer_item_clusters = []
    for p in pointers:
        pointer_cluster = {}
        pointer_cluster_id = 1
        one_token_list = []
        character_per_pointer = [
            character for character in characters_with_pointer
            if character.lower().startswith(p.lower())
        ]
        for character in character_per_pointer:
            character_token = character.split(" ")
            if len(character_token) == 1:
                one_token_list.append(character)
            else:
                found = False
                for key, cluster in pointer_cluster.items():
                    for name in cluster:
                        # print("character",character[len(p):].lower())
                        # print("name in aliases", name[len(p):].lower())
                        if character.endswith("nya") and character[:-3] in name:
                            cluster.append(character)
                            found = True
                            break
                        if character[len(p):].lower() in name[len(p):].lower() or compute_similarity(character[len(p):].lower(),name[len(p):].lower())['jaro_winkler'] >= threshold:

                            print(name[len(p):])
                            print(character[len(p):].lower())
                            print("---")
                            cluster.append(character)
                            found = True
                            break
                if not found:
                    pointer_cluster[pointer_cluster_id] = [character]
                    pointer_cluster_id += 1
        if one_token_list:
            if  len(pointer_cluster) ==1:
                # print(one_token_list)
                # print(pointer_cluster)
                for item in one_token_list:
                    pointer_cluster[1].append(item)
            if not pointer_cluster:
                pointer_cluster[pointer_cluster_id] = one_token_list
        all_pointer_item_clusters.extend(pointer_cluster.values())
    return all_pointer_item_clusters

# Merge clusters
def merge_clusters(aliases_clusters, pointer_clusters):
    final_clusters = {}
    counter = 1
    for cluster in pointer_clusters:
        final_clusters[f"Tokoh-{counter}"] = cluster
        counter += 1
    for key, value in aliases_clusters.items():
        final_clusters[f"Tokoh-{counter}"] = list(value)
        counter += 1
    return final_clusters

# Main clustering function
def cluster_character_aliases(characters_list, pointers):
    aliases_clusters = {}
    characters_with_pointer = [
        character for character in characters_list
        if any(character.lower().startswith(pointer.lower()) for pointer in pointers)
    ]
    print(characters_with_pointer)

    characters_without_pointer = [character for character in characters_list if character not in characters_with_pointer]

    # Cluster characters without pointers
    aliases_clusters = cluster_without_pointers(characters_without_pointer, aliases_clusters, 0.90)
    # Cluster characters with pointers
    pointer_clusters = cluster_with_pointers(characters_with_pointer, pointers, 0.90)
    # Merge all clusters
    return merge_clusters(aliases_clusters, pointer_clusters)

pointers = ['Ibu', 'Pak', 'Puteri', 'Permaisuri', 'Raja', 'Putera', 'Ayah', 'Istri', 'Suami', 'Uwak', 'Menteri', 'Bunda', 'Anak', 'Kakak', 'Adik', 'Kakek', 'Orang Tua', 'Tetangga', 'Putri', 'Beru Tandang', 'Putroe', 'Telangkai', 'Tuhan', 'Abang']
# pointers = []
all_results = []

# Iterate over each row in the DataFrame
for idx, row in sample_df.iterrows():
    story_id = row['story_id']
    sample_data = row['characters']
    
    # Sort aliases by string length (longest first)
    sorted_sample = sorted(set(sample_data), key=len, reverse=True)
    
    # Apply your clustering function
    result = cluster_character_aliases(sorted_sample, pointers)
    
    # Flatten result into a DataFrame-friendly format
    for person, aliases in result.items():
        all_results.append({
            'story_id': story_id,
            'person': person,
            'aliases': aliases
        })

# Create the final DataFrame
df_result = pd.DataFrame(all_results)

```

```python
from ast import literal_eval
from itertools import combinations
import pandas as pd

def get_pairs(clusters, include_singletons=False):
    pairs = set()
    for cluster in clusters:
        cluster = list(cluster)
        if include_singletons and len(cluster) == 1:
            pairs.add((cluster[0], cluster[0]))
        for a, b in combinations(cluster, 2):
            pairs.add(tuple(sorted([a, b])))
    return pairs

def evaluate_clusters(gt_df, pred_df):
    story_ids = gt_df['story_id'].unique()
    
    results = []

    for story_id in story_ids:
        gt_story = gt_df[gt_df['story_id'] == story_id]
        pred_story = pred_df[pred_df['story_id'] == story_id]

        groundtruth_clusters = [
            set(alias.lower() for alias in literal_eval(str(row['aliases'])))
            for _, row in gt_story.iterrows()
        ]

        predicted_clusters = [
            set(alias.lower() for alias in literal_eval(str(row['aliases'])))
            for _, row in pred_story.iterrows()
        ]

        gt_pairs = get_pairs(groundtruth_clusters, include_singletons=True)
        pred_pairs = get_pairs(predicted_clusters, include_singletons=True)
        # print(story_id)
        # print(groundtruth_clusters)
        # print(predicted_clusters)
        # print(gt_pairs & pred_pairs)
        # print("---")
        
        tp = len(gt_pairs & pred_pairs)
        fp = len(pred_pairs - gt_pairs)
        fn = len(gt_pairs - pred_pairs)

        precision = tp / (tp + fp) if (tp + fp) else 0
        recall = tp / (tp + fn) if (tp + fn) else 0
        f1 = 2 * precision * recall / (precision + recall) if (precision + recall) else 0

        results.append({
            'story_id': story_id,
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'tp': tp,
            'fp': fp,
            'fn': fn
        })

    return pd.DataFrame(results)
evaluation_df = evaluate_clusters(ground_truth_df, filtered_df)
evaluation_df

# Optional: overall macro average
print("Average F1:", evaluation_df['f1'].mean())
print("Average precision:", evaluation_df['precision'].mean())
print("Average recall:", evaluation_df['recall'].mean())
```

```python
import pandas as pd
from collections import Counter

# Sample DataFrames (Assuming they are already loaded)

from collections import Counter
def get_character_tag(character_list, tag_dataset, story_id):
    # Filter the tag dataset to include only the relevant story_id
    tag_dataset_filtered = tag_dataset[tag_dataset['story_id'] == story_id]
    tags = []

    # For each character in the list, find its corresponding tag(s) within the relevant story_id
    for character in character_list:
        # Check if the character is in the alias_characters column (converted to list)
        for _, row in tag_dataset_filtered.iterrows():
            # Ensure 'alias_characters' is not NaN before splitting
            if pd.isna(row['final_tag']):
                continue
            
            # Convert alias_characters from string to list by splitting on commas
            alias_list = row['aliases']
            if character.lower() in [alias.lower() for alias in alias_list]:  # Case insensitive matching
                tags.append(row['final_tag'])
                # print(f"Character found: {character} -> Tag: {row['final_tag']}")

    if tags:
        # Return the most common tag
        most_common_tag = Counter(tags).most_common(1)[0][0]
        return most_common_tag
    else:
        return None


# Apply the function to each row of filtered_df and create the new column `final_tag`
filtered_df['final_tag'] = filtered_df.apply(lambda row: get_character_tag(row['aliases'], ground_truth_df, row['story_id']), axis=1)
filtered_df['final_tag'] = filtered_df['final_tag'].apply(lambda x: x if x is not None else "others")
# Show the resulting DataFrame
filtered_df
```
